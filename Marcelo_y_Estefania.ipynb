```python
%%writefile Primer_Punto.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

# Configuraci√≥n de la p√°gina en Streamlit
st.set_page_config(page_title="Espectro de Fourier", page_icon="üìä")
st.title("üìä An√°lisis del Espectro de Fourier")

# Teor√≠a y ecuaciones en LaTeX
st.markdown(r"""## Punto 1

Encuentre la expresi√≥n del espectro de Fourier (forma exponencial y trigonom√©trica) para la se√±al $x(t) = \left|A\cos\left(2\pi*Fo\right)\right|^2$, con $t\in [-1/2Fo, 1/2Fo]$. con A,Fo ‚àà R+. Realice las simulaciones respectivas para graficar el espectro de Fourier del ejercicio 1 (magnitud
 y fase como diagrama de Bode en decibelios), y presente el error relativo y la se√±al reconstruida para N = {1,2,...,50}.



## Soluci√≥n:

Se tiene que:

$$\boxed{{\omega_0} = 2{\pi}F_0}$$

$$x(t)=\left|A\cos\left(2{\pi}F_0t\right)\right|^2 = A^2\cos^2\left(2{\pi}F_0t\right) =  A^2\cos^2\left({\omega_0}t\right)$$

.

recordando que:

$\cos^2(\theta) = \frac{1}{2}\left(1 + cos(2\theta)\right)$

entonces:

$$\boxed{x(t)= \left(\frac{A^2}{2}\right)\left(1 + \cos\left(2{\omega_0}t\right)\right)}$$

Ahora para la serie trigonom√©trica:

$$\boxed{x(t)= a_0 + \sum_{n=1}^{N=50}{a_n\cos(n\omega_0 t) + b_n\sin(n\omega_0 t)}},$$

donde:

$$\boxed{a_0 = c_0 = \frac{1}{t_f-t_i}\int_{t_i}^{t_f}x(t)dt}$$

$$\boxed{a_n = \frac{2}{t_f-t_i}\int_{t_i}^{t_f}x(t)\cos(n\omega_0 t)dt}$$

$$\boxed{b_n = \frac{2}{t_f-t_i}\int_{t_i}^{t_f}x(t)\sin(n\omega_0 t)dt}$$
.

$T = {t_f-t_i} = \frac{1}{2F_0} - (-\frac{1}{2F_0}) = \frac{2}{2F_0}  $

$$\boxed{T = \frac{1}{F_0}}$$


Simulamos la se√±al de entrada con ${F_0} = 10 (Hz)$ ,   $ A = 5$

$$\boxed{x_0(t)= \left(\frac{5^2}{2}\right)\left(1 + \cos\left(2{\pi}(10)t\right)\right)}$$""")


A = 5                                 #Amplitud = A**2
Fo = 10                               #Frecuencia fundamental (Hz)
ti = -1/(2*Fo)                        #Tiempo inicial
tf = 1/(2*Fo)                         #Tiempo final
T = tf-ti                             #Periodo
wo = 2*np.pi*Fo                       #Frecuencia angular
Fs = 100*Fo                      #Frecuencia muestreo estrictamente mayor a 2*Fo
tv = np.arange(ti,tf,1/Fs)            #Vector de tiempo
Nm = len(tv)                          #Muestras en el tiempo
xo = ((A**2)/2)*(1 + np.cos(2*wo*tv)) #se√±al de entrada
fig, ax = plt.subplots()
ax.plot(tv, xo)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("$x(t)$")
ax.set_title("Se√±al Generada")
st.pyplot(fig)



st.markdown(" Dado que $x(t)$ presenta simetr√≠a par, $x(t)=x(-t)$:")

st.latex(r"b_n = 0 \quad \forall n \in \{1,\dots,50\}")

st.markdown("As√≠:")

st.latex(r"x(t)= \left(\frac{A^2}{2} + \frac{A^2}{2}\cos\left(2{\omega_0}t\right)\right) =  a_0 + \sum_{n=1}^{N=50}{a_n\cos(n\omega_0 t)}")

st.markdown(" Calculamos $a_0$ que es quien condiciona el nivel DC de la se√±al.")

st.latex(r"""
a_0 = \frac{1}{\frac{1}{F_0}}\int_{\frac{-1}{2F_0}}^{\frac{1}{2F_0}}
\left(\frac{A^2}{2}\right)\left(1 + \cos\left(2{\omega_0}t\right)\right)dt
""")

st.latex(r"""
a_0= \frac{A^2F_0}{2}\int_{\frac{-1}{2F_0}}^{\frac{1}{2F_0}}(1+\cos(2{\omega_0}t))dt
""")

st.latex(r"a_0 = \frac{A^2}{2}")

st.markdown(" Por otra parte, est√° $a_n$:")

st.latex(r"""
a_n = \frac{2}{\frac{1}{F_0}}\int_{\frac{-1}{2F_0}}^{\frac{1}{2F_0}}
\left(\frac{A^2}{2}\right)\left(1 + \cos\left(2{\omega_0}t\right)\right)(\cos(n{\omega_0}t))dt
""")

st.markdown("Recordando la identidad:")

st.latex(r"\cos(\alpha)\cos(\beta) = \frac{\cos(\alpha+\beta)+\cos(\alpha-\beta)}{2}")

st.latex(r"""
a_n = \left(\frac{A^2}{2{\pi}}\right)\left(\frac{2\sin(n{\pi})}{n}+
\frac{\sin(\left(n+2){\pi}\right)}{(n+2)}+\frac{\sin(\left(n-2){\pi}\right)}{(n-2)}\right)
""")

st.markdown(r"Para $n \neq 0$, $n \neq -2$ y $n \neq 2$, $a_n = 0$. No obstante, para $n = 0$, $n = -2$ y $n = 2$:")


st.latex(r"""
a_n = \left\{
\begin{align}
0 & \quad \forall n \setminus \{-2,2\}\\
\frac{A^2}{2} & \quad n =  \{-2,2\}
\end{align}
\right.
""")

st.latex(r"x(t) = \frac{A^2}{2}+\frac{A^2}{2}\cos(2{\omega_0t})")

st.markdown(" Para el caso de la serie exponencial compleja:")

st.latex(r"c_0 = a_0 = \frac{A^2}{2}")

st.latex(r"c_n = \frac{a_n - j b_n}{2}")

st.latex(r"""
c_n = \left\{
\begin{align}
\frac{A^2}{4} & \quad n = \{-2,2\}\\
\frac{A^2}{2} & \quad n = 0\\
0 & \quad \forall n \setminus \{-2,0,2\}
\end{align}
\right.
""")

st.latex(r"x(t) = \sum_{n=-N}^{N}{c_n e^{j n t}}")

st.latex(r"x(t) = c_{-2}e^{-j2t} + c_{0}e^{0} + c_{2}e^{j2t}")

st.latex(r"""
x(t)= \frac{A^2}{4}(\cos(2{\omega_0t})-j\sin(2{\omega_0t}))+
\frac{A^2}{2}+\frac{A^2}{4}(\cos(2{\omega_0t})+j\sin(2{\omega_0t}))
""")

st.latex(r"x(t) = \frac{A^2}{2}+\frac{A^2}{2}\cos(2{\omega_0t})")


# Se presenta la simulaci√≥n del espectro y la reconstrucci√≥n

# Construir bases serie exponencial
N = 50  # N√∫mero de arm√≥nicos
wo = 2 * np.pi / T  # frecuencia fundamental

# Definir bases
phin = np.zeros((Nm, 2 * N + 1), dtype=np.complex_)  # Crear matriz para guardar bases
for n in range(-N, N + 1, 1):
    phin[:, n + N] = np.exp(1j * n * wo * tv)  # Base de Fourier en el intervalo de inter√©s

# Espectro de la se√±al - serie exponencial
# Calcular espectro -> cambiar seg√∫n la se√±al estudiada
Cn = np.zeros(2 * N + 1, dtype=np.complex_)
nv = np.linspace(-N, N, 2 * N + 1)  # Vector num arm√≥nicos
Cn[N] = (A ** 2) / 2  # Nivel DC C_0
Cn[N - 2] = (A ** 2) / 4  # Nivel arm√≥nico
Cn[N + 2] = (A ** 2) / 4

# Configurar Streamlit
st.title("Simulaci√≥n del Espectro y Reconstrucci√≥n")

# Graficar espectro
fig = plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1)
plt.stem(nv, np.real(Cn), 'r')
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$Re\{C_n\}$', fontsize=14)
plt.axis('tight')
plt.grid()

plt.subplot(2, 2, 2)
plt.stem(nv, np.imag(Cn), 'r')
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$Im\{C_n\}$', fontsize=14)
plt.axis('tight')
plt.grid()

plt.subplot(2, 2, 3)
plt.stem(nv, abs(Cn), 'r')  # Magnitud del espectro
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$|C_n|$', fontsize=14)
plt.axis('tight')
plt.grid()

plt.subplot(2, 2, 4)
plt.stem(nv, np.angle(Cn), 'r')  # Fase del espectro
plt.xlabel(r'$nw_o$[rad/s]', fontsize=14)
plt.ylabel(r'$\langle C_n$', fontsize=14)
plt.axis('tight')
plt.grid()

fig.tight_layout()
st.pyplot(fig)

import streamlit as st
import numpy as np

st.title("C√°lculo del Error Relativo en Fourier")

st.markdown(r"""
Para calcular el error relativo, necesitamos la potencia:

$$e_r[\%] = \frac{\bar{P}_e}{\bar{P_x}} *100[\%] = \left(1-\frac{\sum^N_{n=-N}|c_n|^2P_n}{\bar{P}_x}\right)*100[\%]$$

En este caso, la potencia de $x(t)$ se c√°lcula como:""")

st.latex(r"""\begin{equation}
P_x = \frac{1}{\frac{1}{F_0}}\int\limits_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}{|x(t)^2|dt} = \frac{1}{\frac{1}{F_0}}\int\limits_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}{|\frac{A^2}{2}+\frac{A^2}{2}j\cos(2{\omega_0t})|^2dt}= \frac{A^2F_0}{4}\left(\int\limits_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}dt + \int\limits_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}2*\cos(2{\omega_0t})dt + \int\limits_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}\cos^2(2{\omega_0t})dt\right)
\end{equation}""")

st.markdown(r"""Resolviendo la ecuaci√≥n:

$$\boxed{P_x= \frac{3A^4}{8}}$$

Para el caso de la serie exponencial de Fourier $P_n=1$, entonces:""")

st.latex(r"""
E_r[\%] = 1 - \left(\frac{1}{P_x}\sum\limits_{n=-N}^{N}{|c_n|^2}\right)
""")

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc

# Reconstrucci√≥n
Px = (3*A**4)/(8)

# Figura y axes animaci√≥n
fig, ax = plt.subplots(2, 3, figsize=(12, 5))

# Funci√≥n para gr√°fico interactivo
def update(n):
    ax[0,0].clear()
    ax[0,1].clear()
    ax[0,2].clear()
    ax[1,0].clear()
    ax[1,1].clear()
    ax[1,2].clear()

    # Espectro
    ind = np.arange(N-n, N+n+1)  # Arm√≥nicos a considerar
    ax[0,0].stem(nv, abs(Cn), 'b')  # Magnitud del espectro
    ax[0,0].stem(nv[ind], abs(Cn[ind]), 'r')  # Magnitud del espectro
    ax[0,0].set_xlabel(r'$nw_o$[rad/s]')
    ax[0,0].set_ylabel(r'$|C_n|$')

    # Componentes espectrales
    ce = phin[:, ind].dot(np.diag(Cn[ind]))
    ax[0,1].plot(np.real(ce))  # Componente real
    ax[0,1].set_xlabel('t[s]')
    ax[0,1].set_ylabel(r'$Re\{c_n\phi_n(t)\}$')

    ax[0,2].plot(np.imag(ce))  # Componente imaginario
    ax[0,2].set_xlabel('t[s]')
    ax[0,2].set_ylabel(r'$Im\{c_n\phi_n(t)\}$')

    # Reconstrucci√≥n
    er = 1 - np.sum(abs(Cn[ind])**2) / Px
    xe = phin[:, ind].dot(Cn[ind])  # Se√±al reconstruida
    et = xo - xe  # Error en t

    ax[1,1].plot(tv, xo, 'r', linewidth=4, label='$x(t)$')
    ax[1,1].plot(tv, xe, color='b', label='$x_e(t)$')
    ax[1,1].plot(tv, et, color='g', label='$e(t)$')
    ax[1,1].set_title(r"$n=%d. E_r=%.2f\%%$" % (n, 100 * er))

    ax[1,1].set_xlabel('t[s]')
    ax[1,1].set_ylabel('x(t)')
    ax[1,1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax[1,0].axis('off')
    ax[1,2].axis('off')
    plt.subplots_adjust(wspace=0.5, hspace=0.7)

    st.pyplot(fig)  # Mostrar la figura en Streamlit

# UI en Streamlit
st.title("Animaci√≥n de Reconstrucci√≥n Espectral")
n = st.slider("Selecciona n", 0, N, 1)  # Control deslizante para ajustar n
update(n)
st.success(" Punto 1 completado")






```

```python
%%writefile Tercer_Punto.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.signal as sig
import scipy.optimize as opt
import sympy as sym



st.set_page_config(page_title="¬øQu√© es la distorsi√≥n total de arm√≥nicos? ¬øEn qu√© consiste el factor de potencia en un circuito el√©ctrico? ", page_icon="‚ö°")
st.title("¬øQu√© es la distorsi√≥n total de arm√≥nicos? ¬øEn qu√© consiste el factor de potencia en un circuito el√©ctrico? Entiende mejor con un ejemplo interactivo üòÄ‚ö°")

st.write(r"""Consulte en qu√© consiste la distorsi√≥n total de arm√≥nicos (*Total Harmonic Distortion*-(THD) y el factor de potencia en un circuito el√©ctrico. C√≥mo puede calcularse el THD desde la FFt? Genere un ejemplo ilustrativo para el c√°lculo del THD y la distorsi√≥n del factor de potencia para un rectificador de onda completa con carga: i)netamente resistiva y ii) carga RC en serie. Establezca las condiciones necesarias para las simulaciones. El usuario podr√° escoger diferentes valores de R y C. Discuta los valores obtenidos.""")
st.divider()
st.header("Soluci√≥n")
st.write("""La distorsi√≥n arm√≥nica total (THD) es una medida que cuantifica la distorsi√≥n en una forma de onda de voltaje o corriente alterna (CA) debido a la presencia de arm√≥nicos. Los arm√≥nicos son frecuencias que son m√∫ltiplos enteros de la frecuencia fundamental y pueden distorsionar la forma de onda original, creando picos y valles que no est√°n presentes en una onda sinusoidal pura

El objetivo de realizar un ejemplo ilustrativo para el c√°lculo del **THD (Total Harmonic Distortion o Distorsi√≥n Arm√≥nica Total)** y la **distorsi√≥n del factor de potencia** para un rectificador de onda completa con carga es comprender c√≥mo estas caracter√≠sticas afectan la calidad y eficiencia de los sistemas de conversi√≥n de energ√≠a. Esto es especialmente importante en aplicaciones de electr√≥nica de potencia, donde los rectificadores son com√∫nmente utilizados para convertir corriente alterna (AC) en corriente continua (DC).""")

st.subheader("Distorsi√≥n arm√≥nica total (THD):")

st.write("""El **THD** es una medida de la distorsi√≥n en una se√±al debido a la presencia de arm√≥nicos, los cuales son m√∫ltiplos de la frecuencia fundamental. Cuanto mayor sea el THD, m√°s alejada est√° la se√±al de salida de una forma de onda ideal, lo que afecta el rendimiento de los equipos conectados.

En los sistemas de rectificaci√≥n, la presencia de arm√≥nicos es com√∫n debido a la no linealidad del proceso de rectificaci√≥n. Esta simulaci√≥n permite al usuario ver c√≥mo los diferentes valores de los componentes (resistencias y capacitancias) afectan el contenido arm√≥nico de la se√±al rectificada y, en consecuencia, el THD.
se calcula mediante:""")
st.latex(r"""
\text{THD (\%)} = \frac{\sqrt{P_2 + P_3 + P_4 + \cdots}}{P_1} \times 100
= \frac{\sqrt{V_2^2 + V_3^2 + \dots + V_n^2}}{V_1} \times 100
""")

st.markdown("""
Donde:
- \( P_1 \) es la potencia de la frecuencia fundamental.
- \( P_2, P_3, ..., P_n \) es la potencia para el arm√≥nico \( n \) de la fundamental.
- Cada \( V_i \) corresponde a la amplitud respectiva del arm√≥nico.
""")
st.subheader("Distorci√≥n del factor de potencia (DFP):""")

st.write("""El factor de potencia mide cu√°n eficiente es el uso de la energ√≠a el√©ctrica en un sistema. Se calcula como la relaci√≥n entre la potencia activa y la potencia aparente. La **distorsi√≥n del factor de potencia** indica c√≥mo los arm√≥nicos impactan en la eficiencia de la transmisi√≥n de energ√≠a.

En cargas puramente resistivas, se espera que el factor de potencia est√© m√°s cerca de 1, lo que indica una conversi√≥n eficiente de la potencia. Sin embargo, en cargas inductivas o capacitivas, la distorsi√≥n del factor de potencia puede aumentar, lo que indica que hay m√°s energ√≠a reactiva en el sistema que no contribuye a la potencia √∫til.""")

st.divider()
st.header("Simulaci√≥n")
s=sym.symbols('s',complex=True)
t,R,L,C=sym.symbols('t,R,L,C', positive=True) #Definimos variables
X=sym.Function("X")(s)
Y=1/(R*C*s+1)*X #Funci√≥n de transferencia
st.title("An√°lisis de circuitos RC Y R Puro")
circuito= st.radio("Seleccione el tipo de circuito",["Circuito RC serie","Circuito R puro"])
#Las entradas las podr√° elegir el ususario:
if circuito == "Circuito RC serie":
  R_v = st.number_input("Ingrese el valor de R (Œ©):", min_value=0.0, value=1000.0)

# Entrada de capacitancia con selecci√≥n de unidad
  cap_value = st.number_input("Ingrese el valor de C:", min_value=0.0, value=100.0)
  cap_unit = st.selectbox("Unidad de C:", ["F", "mF", "¬µF", "nF", "pF"])

# Convertir la capacitancia a Faradios
  unit_conversion = {"F": 1, "mF": 1e-3, "¬µF": 1e-6, "nF": 1e-9, "pF": 1e-12}
  C_v = cap_value * unit_conversion[cap_unit]

if circuito == "Circuito RC serie":
    Y_RC = Y.subs(R, R_v).subs(C, C_v)
    st.write(f"La respuesta del circuito RC es: {Y_RC}")

elif circuito == "Circuito R puro":
    R_v = st.number_input("Ingrese el valor de R (Œ©):", min_value=0.0, value=1000.0)# No hay capacitor en el circuito R puro
    C_v=0.0
    Y_R = Y.subs(R, R_v).subs(C,C_v)
    st.write(f"La respuesta del circuito R puro es: {Y_R}")

else:
  st.error("Opci√≥n no v√°lida")
#Funci√≥n de transferencia en scipy
num=np.array([1]) #coeficientes del numerador de la funcion de transferencia como numpy array
den=np.array([R_v*C_v,1])#coeficientes del denominador de la funcion de transferencia como numpy array
G_n=sig.TransferFunction(num,den)

#Definimos los par√°metros de la se√±al
A=st.number_input("Ingrese la amplitud de la se√±al:",min_value=0.0,value=1.0)#Amplitud de entrada
Fo=st.number_input("Ingrese la frecuencia de la se√±al (Hz):",min_value=1.0,value=1000.0) #Frecuencia de alimentaci√≥n
fs=100*Fo #Frecuencia e muestreo
To=1/Fo #Periodo fundamental
Ts=1/fs #Frecuencia fundamental
t=np.arange(0,7*To,Ts) # se simulan 5 peridos de alimentaci√≥n con un muestreo segun Ts
in_o=A*(np.sin(2*np.pi*Fo*t))
rec_c=sig.square(2*np.pi*Fo*t)#funci√≥n tren de pulsos para simular se√±al rectificada
in_=in_o*rec_c
#Evaluamos la salida
out=G_n.output(in_,T=t)[1]
#Procedemos a graficar
fig, ax = plt.subplots()
ax.plot(t, in_o, label='Alimentaci√≥n')
ax.set_xlabel('Tiempo [s]')
ax.legend()
st.pyplot(fig)

fig, ax = plt.subplots()
ax.plot(t, rec_c, label="Tren de pulsos")
ax.legend()
st.pyplot(fig)

fig, ax = plt.subplots()
ax.plot(t, in_, label='|In(t)|')  # Entrada rectificada
ax.plot(t, out, label='Out(t)')  # Salida del circuito
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('out(t)')
ax.legend()
st.pyplot(fig)
#Calculamos los espectros
vfre=np.fft.rfftfreq(len(out),1/fs)#vector de frecuencia en Hz segun tama√±o del vector y tiempo de muestreo
Xf=np.fft.rfft(out)
magXf=np.abs(Xf)
V_n=(2*magXf)/len(out)

#Calculamos la THD
V2=[0 if i==0 or i==1 else V_n[i]**2 for i in range(len(V_n))]
THD=np.sqrt(sum(V2))/V_n[1] if circuito== "Circuito RC serie" else 0 #Se calcula la THD solo si es un circuito RC
THD=round(THD,4)
#THD mide cu√°nta energ√≠a hay en los arm√≥nicos superiores en comparaci√≥n con la fundamental.

#Factor de potencia
PF_thd_=np.sqrt(1/(1+(THD**2))) #Calcula el factor de potencia considerando el factor de potencia
fig, ax = plt.subplots()
ax.stem(vfre, abs(Xf / len(out)))
ax.set_xlabel('Frecuencia [Hz]')
st.pyplot(fig)

st.write(f"**THD = {THD} %**")
st.write(f"**DPF = {PF_thd_}**")
st.subheader("An√°lisis de los resultados:")
st.write("""Se observa como a medida que variamos los valores de R y C el THD no variar√°. Una resistencia alta o una capacitancia baja tienden a aumentar la distorci√≥n arm√≥nica.
Un THD alto lleva a un factor de potencia mas bajo indicando una menor eficiencia. En conclusi√≥n ajustar los valores de R y C nos permiten mejorar la eficiencia minimizando distorciones.""")
st.success(" Punto 3 completado, el punto 4 lo puedes encontar en:")
st.markdown("https://github.com/estefaniabrito/SYS/blob/main/Punto_4.ipynb")





```
